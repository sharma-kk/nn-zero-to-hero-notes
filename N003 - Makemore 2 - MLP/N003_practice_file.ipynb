```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
words = open('../names.txt', 'r').read().splitlines()
print(words[:8])
print(len(words))
```

```python
chars = sorted(list(set(''.join(words))))

stoi = {s: i+1 for i,s in enumerate(chars)}
stoi['.'] = 0

itos = {i:s for s,i in stoi.items()}

print(itos)
print()
print(stoi)
```

```python
block_size = 3

X, Y = [], []

for w in words[:5]:
    print(f'\n{w}')
    context = [0] * block_size
    for ch in w + '.':
        ix = stoi[ch]
        X.append(context)
        Y.append(ix)
        print(''.join(itos[i] for i in context), '--->', itos[ix])
        context = context[1:] + [ix]

X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
# This is what the dataset looks like right now
print(f'Input:  {X.shape}\t{X.dtype}\t{X[4]}') # 32 times the 3 character indexes before the next one
print(f'Output: {Y.shape}\t{Y.dtype}\t{Y[5]}') # 32 times the next character's index
```
